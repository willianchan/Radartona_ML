```python
for index, row in df_clima.iterrows():
    main = row["weather"][0]["main"]
    if main == "Rain":
        r = 1
    else:
        r = 0
        
    df_clima.loc[index, 'rain'] = r
```

```python
df_clima.head()
```

```python
for index, row in df_clima.iterrows():
    ano, mes, dia = list(map(int, row["dt_iso"][:10].split("-")))
    
    df_clima.loc[index, 'ano'] = ano
    df_clima.loc[index, 'mes'] = mes
    df_clima.loc[index, 'dia'] = dia
```

```python
df_clima['dia'] = df_clima['dia'].astype(int)
```
