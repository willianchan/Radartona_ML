```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/Willian/Desktop/Radartona/bases_tratadas/base_radares.csv', sep=";")
```

```python
df.head()
```

```python
df_fim = pd.DataFrame(columns=['id_radar','velocidade_maxima'])

for index, row in df.iterrows():
    ids = row['id_radar'].replace(" ","").split("-").split("/")
    for id in ids:
#         df2 = pd.DataFrame([id, row['velocidade_maxima']])
#         df2 = pd.DataFrame({"id_radar":[id],"velocidade_maxima":[row['velocidade_maxima']]})
#         print(df2)
        df_fim = df_fim.append({"id_radar": str(id), "velocidade_maxima": row["velocidade_maxima"]}, ignore_index=True)
#     df.loc[index, 'bairro'] = bairro
```

```python
df_fim
```

```python
for index, row in df_fim.iterrows():
    velocidade = str(row['velocidade_maxima']).replace(" km/h", "")
    df_fim.loc[index,"velocidade_maxima"] = velocidade
```

```python
df_fim
```

```python
df_fim.loc[df_fim['id_radar'] == '4651', "velocidade_maxima"] = '50'
df_fim.loc[df_fim['id_radar'] == '4650', "velocidade_maxima"] = '50'
```

```python
for index, row in df_fim.iterrows():
    velocidade = list(map(float,str(row['velocidade_maxima']).split('/')))
    v = sum(velocidade)/len(velocidade)
    df_fim.loc[index,"velocidade_maxima"] = int(v)
```
